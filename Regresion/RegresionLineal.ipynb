# Análisis Exploratorio de Datos (EDA) y Evaluación del Modelo de Regresión Lineal

## 1. Inspección de los Datos

- **Eliminación de valores nulos**: Se eliminaron los valores nulos mediante `df.dropna()` para garantizar que el conjunto de datos estuviera limpio.
- **Creación de la variable objetivo**: La variable objetivo `Total_Spending` se creó sumando los gastos en diferentes productos como `MntWines`, `MntFruits`, `MntMeatProducts`, entre otros.
- **Codificación de variables categóricas**: Las variables categóricas como `Education`, `Marital_Status` y `Dt_Customer` fueron transformadas a variables numéricas utilizando `LabelEncoder()` de `sklearn`.
- **Transformación logarítmica**: Las variables `MntWines` y `Total_Spending` fueron transformadas usando `np.log1p()` para reducir el impacto de los valores extremos y mejorar la normalidad de los datos.

## 2. Distribución de Datos

- Se utilizó un **heatmap de correlación** para identificar las relaciones entre las variables. 
- La variable `MntWines` mostró la mayor correlación con `Total_Spending`, lo que explica su alto coeficiente en el modelo de regresión lineal.

## 3. Multicolinealidad y VIF

- Se calculó el **Variance Inflation Factor (VIF)** para verificar la presencia de colinealidad entre las variables. 
- Si el VIF de alguna variable supera el valor de 10, se considera que hay una fuerte colinealidad, lo que podría implicar que se elimine alguna de las variables correlacionadas para mejorar la estabilidad del modelo.

## 4. Normalización y División de Datos

- Se utilizó **`StandardScaler()`** para escalar las variables predictoras y garantizar que todas las características tuvieran una escala similar, lo que es fundamental para mejorar el desempeño del modelo de regresión.
- Los datos fueron divididos en **80% de entrenamiento y 20% de prueba** utilizando `train_test_split()`. Esto permite evaluar el rendimiento del modelo en datos no vistos.

## 5. Desempeño del Modelo

Las siguientes métricas fueron utilizadas para evaluar el desempeño del modelo de regresión lineal:

| **Métrica** | **Valor** |
|-------------|-----------|
| **R²**      | 0.957     |
| **MSE**     | 0.10      |
| **MAE**     | 0.23      |
| **RMSE**    | 0.31      |

- **R² = 0.957**: Esto indica que el modelo es capaz de explicar el **95.7%** de la variabilidad del gasto total, lo cual es un excelente indicador de su precisión.
- **MSE (Error Cuadrático Medio)** y **MAE (Error Absoluto Medio)** son relativamente bajos, lo que confirma que el modelo está realizando buenas predicciones.
- **RMSE (Raíz del Error Cuadrático Medio)** también es bajo, lo que refuerza la idea de que las predicciones son precisas.

## 6. Análisis de Residuos

- **Distribución de residuos**: La distribución de los residuos muestra un comportamiento normal, lo cual sugiere que los errores de predicción están distribuidos de manera aleatoria y no existe un sesgo sistemático.
- **Valores reales vs. predichos**: La mayoría de los puntos en el gráfico de dispersión siguen la línea diagonal, lo que indica que las predicciones son muy precisas y los valores reales y predichos están bastante alineados.

## 7. Importancia de las Variables

- **MntWines**: Esta variable sigue siendo la más influyente en la predicción del gasto total, lo que se refleja en su alto coeficiente en el modelo de regresión.
- **Otras variables relevantes**: Otras categorías de gasto, como `MntMeatProducts`, `MntGoldProds`, entre otras, también tienen un impacto considerable en la predicción del gasto total.

## 8. Conclusión

El modelo de regresión lineal muestra una alta precisión con un **R² de 0.957**, lo que sugiere que es capaz de explicar una gran parte de la variabilidad en el gasto total de los clientes. La distribución de residuos y la comparación entre valores reales y predichos indican que el modelo generaliza bien y hace buenas predicciones. En resumen, el modelo es robusto y útil para predecir el comportamiento de gasto total en base a las características de los clientes.

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from statsmodels.stats.outliers_influence import variance_inflation_factor
import matplotlib.pyplot as plt
from io import StringIO

# # Ruta del archivo cargado
# file_path = "../marketing_campaign.csv"

# # Reemplazar espacios por comas y cargar el archivo en un DataFrame
# with open(file_path, "r", encoding="utf-8") as file:
#     content = file.read().replace("\t", ",")

# # Guardar el contenido corregido
# corrected_file_path = "marketing_campaign_corrected.csv"
# with open(corrected_file_path, "w", encoding="utf-8") as file:
#     file.write(content)

# Cargar el dataset corregido en un DataFrame de Pandas
df = pd.read_csv("marketing_campaign_corrected.csv")

# Mostrar las primeras filas para inspección
df.head()

# Eliminar valores nulos
df = df.dropna()

print(df.columns.to_list())
df.columns = df.columns.str.strip()

# Crear la columna Total_Spending
gasto_cols = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['Total_Spending'] = df[gasto_cols].sum(axis=1)

# Codificación de variables categóricas
categorical_cols = ['Education', 'Marital_Status', 'Dt_Customer']
label_encoder = LabelEncoder()
for col in categorical_cols:
    df[col] = label_encoder.fit_transform(df[col])

# Calcular VIF para todas las variables predictoras
X_vif = df.drop(columns=['Total_Spending'])
vif_data = pd.DataFrame()
vif_data["Variable"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(len(X_vif.columns))]

print(vif_data)

df['MntWines_log'] = np.log1p(df['MntWines'])  # Logaritmo para reducir impacto de valores extremos
df['Total_Spending_log'] = np.log1p(df['Total_Spending'])

# Definir variables predictoras y objetivo
X = df.drop(columns=['ID', 'Total_Spending', 'Total_Spending_log'])
y = df['Total_Spending_log']

plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Matriz de Correlación")
plt.show()

# Normalización de los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# División de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Crear y entrenar el modelo
model = LinearRegression()
model.fit(X_train, y_train)

# Predicción
y_pred = model.predict(X_test)

residuals = y_test - y_pred

plt.figure(figsize=(8, 5))
sns.histplot(residuals, bins=30, kde=True)
plt.axvline(x=0, color='r', linestyle='--')
plt.title("Distribución de los residuos")
plt.xlabel("Error de Predicción")
plt.ylabel("Frecuencia")
plt.show()

coeff_df = pd.DataFrame({
    'Característica': X.columns,
    'Coeficiente': model.coef_
}).sort_values('Coeficiente', ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x='Coeficiente', y='Característica', data=coeff_df, palette='viridis')
plt.title('Importancia de las Características en la Predicción')
plt.show()

# Evaluación del modelo
print(f'R²: {r2_score(y_test, y_pred):.3f}')
print(f'MSE: {mean_squared_error(y_test, y_pred):.2f}')
print(f'MAE: {mean_absolute_error(y_test, y_pred):.2f}')
print(f'RMSE: {np.sqrt(mean_squared_error(y_test, y_pred)):.2f}')

# Visualización de coeficientes
coeff_df = pd.DataFrame({'Característica': X.columns, 'Coeficiente': model.coef_}).sort_values('Coeficiente', ascending=False)
print("\nCoeficientes del modelo:")
print(coeff_df)

# Graficar valores reales vs predichos
plt.figure(figsize=(10, 5))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.6)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.xlabel('Gasto Real')
plt.ylabel('Gasto Predicho')
plt.title('Valores Reales vs Predichos')
plt.show()
```
